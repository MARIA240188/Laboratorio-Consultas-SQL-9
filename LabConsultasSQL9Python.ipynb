```python
import pymysql
# Establecer la conexión a la base de datos
connection = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='1234',
    database='sakila'
)
```

```python
query_may = "SELECT customer_id, COUNT(*) AS rentals_in_may FROM rentals_may GROUP BY customer_id;"
query_june = "SELECT customer_id, COUNT(*) AS rentals_in_june FROM rentals_june GROUP BY customer_id;"
```

```python
df_may = pd.read_sql(query_may, connection)
df_june = pd.read_sql(query_june, connection)
```

```python
# 8. Escriba una función que verifique si el cliente tomó prestadas más o menos películas en el mes de junio en comparación con mayo.
```

```python
def compare_rentals(df_may, df_june):
    merged_df = pd.merge(df_may, df_june, on='customer_id', suffixes=('_may', '_june'))
    merged_df['comparison'] = 'Equal'
    merged_df.loc[merged_df['rentals_in_june'] > merged_df['rentals_in_may'], 'comparison'] = 'More in June'
    merged_df.loc[merged_df['rentals_in_june'] < merged_df['rentals_in_may'], 'comparison'] = 'Less in June'
    return merged_df

comparison_df = compare_rentals(df_may, df_june)
print(comparison_df)
```
